```python
def draw_widgets():
    #quantity_graph = widgets.IntSliderWidget()
    quantity_graph.min= 1
    quantity_graph.max= 20
    quantity_graph.value = 1
    quantity_graph.description = 'Number of graphs'
    
    #choise_feature_dropdown = widgets.DropdownWidget()
    #choise_feature_dropdown.values  = feature
    #choise_feature_dropdown.value = 0
    #choise_feature_dropdown.description = 'Select feature for research'
    
    #choise_graph_dropdown = widgets.DropdownWidget()
    #choise_graph_dropdown.values  = type_graph
    #choise_graph_dropdown.value = 0
    #choise_graph_dropdown.description = 'Select which type of graph'
    
    #driver_number_text_input = widgets.TextWidget()
    driver_number_text_input.description = 'The driver number'
    driver_number_text_input.value = '1' 
    
    display(quantity_graph)
    #display(choise_feature_dropdown)
    #display(choise_graph_dropdown)
    display(driver_number_text_input)
    

def on_trait_change():
    print "Change", driver_number_text_input.value
    choise_driver(driver_number_text_input.value, choise_graph_dropdown.value)

def draw_graph_all(data, v, a, cur_pos):
    pos = cur_pos
    #plt.close('all')
    pd.set_option('display.mpl_style', 'default') 
    #pd.set_option('display.line_width', 4500)
    pd.get_option("display.max_rows")
    pd.get_option("display.max_columns")
    fig = plt.figure(figsize(10, 10), dpi=100)
    fig, axes = plt.subplots(figsize=(15,10))
    axes.set_xlabel('x')
    axes.set_ylabel('y')
    ax = 'Length {}'.format(pos)
    axes.set_title(ax)
    axes.plot(data[pos][:, 0], data[pos][:, 1], 'g')
        
    pd.set_option('display.mpl_style', 'default') 
    #pd.set_option('display.line_width', 4500) 
    pd.get_option("display.max_rows")
    pd.get_option("display.max_columns")
    fig = plt.figure(figsize(10, 10), dpi=100)
    fig, axes = plt.subplots(figsize=(15,10))
    axes.set_xlabel('x')
    axes.set_ylabel('y')
    ax = 'Velocity/Acceleration {}'.format(pos)
    axes.set_title(ax)
    axes.plot(v, 'r')
    axes.plot(a, 'b')
       
    #plt.close('all')

def draw_graph(data, cur_pos):
    pos = cur_pos-10
    while pos < cur_pos:
        
        v, v_mean, v_median, v_std = calculate_driver_velocitys(data[pos])
        a, a_mean, a_median, a_std = calculate_driver_acceleration(v)
        draw_graph_all(data, v, a, pos)
        draw_inta(v, a, pos)
        pos += 1
    

def draw_inta(v, a, pos):
    pd.set_option('display.mpl_style', 'default') 
    #pd.get_option("display.max_rows")
    fig = plt.figure(figsize(10, 10), dpi=100)
    fig, axes = plt.subplots(figsize=(15,10))
    ax = 'Velocity/Acceleration {}'.format(pos)
    axes.set_title(ax)
    axes.hist(v)
    axes.hist(a)
```

```python

def calculate_driver_acceleration(v):
    a = np.diff(v)
    return a, np.mean(a), np.median(a), np.std(a)


def calculate_driver_velocitys(data):
    #data_n = preprocessing.normalize(data)
    v = np.hypot(np.diff(data[:, 0]), np.diff(data[:, 1]))
    #v_n = preprocessing.normalize(v)
    return v, np.mean(v), np.median(v), np.std(v)

def choise_driver(driver, feature):
    path = os.path.join(DRIVERS_DATA_DIR, driver, '{}.npy'.format(driver))
    if os.path.exists(path):
        data = np.load(path)
        plt.close('all')
        draw_graph(data, quantity_graph.value*10)
    else:
        print('File {} not found'.format(path))
        
        
```
