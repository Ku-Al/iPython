```python
"""
def draw_graph_all(data, driver):
    
    m_l, m_m_l, m_std_l, m_v, m_m_v, m_std_v, m_a, m_m_a, m_std_a = calculate_mean_l_v_a(data)
 
    pd.set_option('display.mpl_style', 'default') 
    fig, axes = plt.subplots(figsize=(15,10))
    axes.set_xlabel('x')
    axes.set_ylabel('y')
    ax = 'Length {}'.format(driver)
    axes.set_title(ax)
    x = [x for x in xrange(0,200)]
    axes.scatter(x, m_l, s=20, c='g', marker='o')
    axes.axhline(y=m_m_l, c='r')
    axes.axhline(y=m_std_l, c='y')
    y=m_m_l*3
    axes.axhline(y, c='r')
        
    pd.set_option('display.mpl_style', 'default') 
    fig = plt.figure(figsize(10, 10), dpi=100)
    fig, axes = plt.subplots(figsize=(15,10))
    axes.set_xlabel('x')
    axes.set_ylabel('y')
    ax = 'Velocity {}'.format(driver)
    axes.set_title(ax)
    axes.scatter(x, m_v, s=20, c='g', marker='o')
    axes.axhline(y=m_m_v, c='r')
    y=m_m_v+m_std_v*3
    axes.axhline(y, c='b')
    y=m_m_v-m_std_v*3
    axes.axhline(y, c='b')
    
    pd.set_option('display.mpl_style', 'default') 
    fig = plt.figure(figsize(10, 10), dpi=100)
    fig, axes = plt.subplots(figsize=(15,10))
    axes.set_xlabel('x')
    axes.set_ylabel('y')
    ax = 'Acceleration {}'.format(driver)
    axes.set_title(ax)
    axes.scatter(x, m_a, s=20, c='g', marker='o')
    axes.axhline(y=m_m_a, c='r')
    y = m_m_a+m_std_a*3
    axes.axhline(y)
    y=m_m_a-m_std_a*3
    axes.axhline(y)
"""
    
def calculate_driver_acceleration(v):
    a = np.diff(v)
    return a, np.mean(a), np.median(a), np.std(a)


def calculate_driver_velocitys(data):
    v = np.hypot(np.diff(data[:, 0]), np.diff(data[:, 1]))
    return v, np.mean(v), np.median(v), np.std(v)

def calculate_mean_l_v_a(data):
    m_vel = []
    m_acc = []
    m_len = []
    i=0
    while i < 200:
        v, v_mean, v_median, v_std = calculate_driver_velocitys(data[i])
        m_vel.append(v_mean)
        m_len.append(sum(v))
        a, a_mean, a_median, a_std = calculate_driver_acceleration(v)
        m_acc.append(a_mean)
        i += 1
    z_m_vel = stats.zscore(m_vel)
    z_m_acc = stats.zscore(m_acc)
    return m_len, np.mean(m_len), np.median(m_len), z_m_vel, np.mean(z_m_vel), \
           np.std(z_m_vel), z_m_acc, np.mean(z_m_acc), np.std(z_m_acc)


def choise_driver(driver):
    path = os.path.join(DRIVERS_DATA_DIR, driver, '{}.npy'.format(driver))
    if os.path.exists(path):
        data = np.load(path)    
        ##draw_graph_all(data, driver)
    else:
        print('File {} not found'.format(path))
    return data        
```

```python
driver_number_text_input = widgets.TextWidget()
draw_widgets()
driver_number_text_input.on_trait_change(on_trait_change, 'value')
data = choise_driver(driver_number_text_input.value)

m_l, m_m_l, m_std_l, m_v, m_m_v, m_std_v, m_a, m_m_a, m_std_a = calculate_mean_l_v_a(data)
```

```python
m_v = m_v/LA.norm(m_v)
print "scale"
print m_v
```

```python
X = PCA(n_components=2).fit_transform(m_v)
d_l_num = X.shape[0]
OUTLIER_FRACTION = 0.09
```

```python
print X
print "-->", d_l_num
```

```python
clf = svm.OneClassSVM(kernel="rbf")
clf.fit(X)
```

```python

dist_to_border = clf.decision_function(X).ravel()
threshold = stats.scoreatpercentile(dist_to_border,
            100 * OUTLIER_FRACTION)
is_inlier = dist_to_border > threshold
```

```python
xx, yy = np.meshgrid(np.linspace(-7, 7, 500), np.linspace(-7, 7, 500))
n_inliers = int((1. - OUTLIER_FRACTION) * d_l_num)
n_outliers = int(OUTLIER_FRACTION * d_l_num)
Z = clf.decision_function(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)
plt.title("Outlier detection")
plt.contourf(xx, yy, Z, levels=np.linspace(Z.min(), threshold, 7),
                         cmap=plt.cm.Blues_r)
a = plt.contour(xx, yy, Z, levels=[threshold],
                            linewidths=2, colors='red')
plt.contourf(xx, yy, Z, levels=[threshold, Z.max()],
                         colors='orange')
b = plt.scatter(X[is_inlier == 0, 0], X[is_inlier == 0, 1], c='white')
c = plt.scatter(X[is_inlier == 1, 0], X[is_inlier == 1, 1], c='black')
plt.axis('tight')
plt.legend([a.collections[0], b, c],
           ['learned decision function', 'outliers', 'inliers'],
           prop=matplotlib.font_manager.FontProperties(size=11))
plt.xlim((-7, 7))
plt.ylim((-7, 7))
plt.show()
```
